```python
# ---------- 1) PRINT SHAPES + DISTRIBUTION ----------
def print_dataset_info(datasets, target_name="Annotations", sort_index_labels=True):
    """
    datasets: dict like {"risto": (X_risto, y_risto), ...}
    """
    for name, (X, y) in datasets.items():
        print(f"\n=== {name.upper()} ===")
        print("X shape:", X.shape)
        print("y shape:", y.shape)
        print("Target distribution:")
        vc = y.value_counts()
        if sort_index_labels:
            vc = vc.sort_index()   # да е секогаш 0 па 1
        print(vc)


# ---------- helpers for SMART DESCRIBE ----------
def zero_crossing_rate(series):
    x = series.dropna().to_numpy()
    return np.mean(x[:-1] * x[1:] < 0) if len(x) > 1 else 0


# ---------- 2) SMART DESCRIBE FOR ONE X (features df) ----------
def zero_crossing_rate(series):
    x = pd.to_numeric(series, errors="coerce").dropna().to_numpy()
    return np.mean(x[:-1] * x[1:] < 0) if len(x) > 1 else 0

def safe_skew(s):
    x = pd.to_numeric(s, errors="coerce").dropna()
    if x.nunique() <= 1:   # константна/скоро константна
        return 0.0
    return skew(x.to_numpy())

def safe_kurtosis(s):
    x = pd.to_numeric(s, errors="coerce").dropna()
    if x.nunique() <= 1:
        return 0.0
    return kurtosis(x.to_numpy())

def smart_describe_X(X, sort_by="rms", ascending=False):
    # само нумерички, и чистење inf
    Xn = X.apply(pd.to_numeric, errors="coerce").replace([np.inf, -np.inf], np.nan)

    mean = Xn.mean()
    std  = Xn.std()

    desc = pd.DataFrame({
        "mean": mean,
        "std": std,
        "cv": std / (mean.abs() + 1e-6),
        "min": Xn.min(),
        "max": Xn.max(),
        "skew": Xn.apply(safe_skew),
        "kurtosis": Xn.apply(safe_kurtosis),
        "rms": np.sqrt((Xn.fillna(0)**2).mean()),   # fillna за да не пука sqrt
        "energy": (Xn.fillna(0)**2).sum(),
        "zero_crossing_rate": Xn.apply(zero_crossing_rate),
        "n_nan": Xn.isna().sum(),                   # BONUS: колку NaN има по колона
        "n_unique": Xn.nunique(dropna=True)         # BONUS: дали е константна
    })

    if sort_by is not None:
        desc = desc.sort_values(sort_by, ascending=ascending)

    return desc


# ---------- 3) SMART DESCRIBE FOR ALL PEOPLE IN datasets ----------
def smart_describe_datasets(datasets, sort_by="rms", ascending=False):
    """
    datasets: dict {"risto": (X_risto, y_risto), ...}
    враќа dict {"risto": desc_df, ...}
    """
    out = {}
    for name, (X, y) in datasets.items():
        out[name] = smart_describe_X(X, sort_by=sort_by, ascending=ascending)
    return out
```

```python
smart_stats = smart_describe_datasets(datasets, sort_by="rms", ascending=False)
```

```python
from IPython.display import display

def show_smart_stats_all(datasets, sort_by="rms", ascending=False):
    smart_stats = smart_describe_datasets(datasets, sort_by=sort_by, ascending=ascending)
    for person, df_stats in smart_stats.items():
        print(f"\n=== {person.upper()} ===")
        display(df_stats)
    return smart_stats

smart_stats = show_smart_stats_all(datasets)
```

```python
def split_by_time_gaps(df, time_col="Hardware Timestamp", gap_factor=50):
    df = df.sort_values(time_col).reset_index(drop=True)
    dt = df[time_col].diff()
    thr = dt.median() * gap_factor
    segments = (dt > thr).cumsum()
    return [seg.reset_index(drop=True) for _, seg in df.groupby(segments)]
```

```python
def rolling_rms(x, win):
    x = np.asarray(x, dtype=float)
    return np.sqrt(np.convolve(x**2, np.ones(win)/win, mode="same"))

def shade_annotation_ones(ax, t, y_bin, alpha=0.18):
    """
    t: 1D array time (seconds) for the same indices as y_bin
    y_bin: 0/1 array
    """
    y_bin = np.asarray(y_bin).astype(int)
    t = np.asarray(t)

    in_one = False
    start_t = None

    for i in range(len(y_bin)):
        if y_bin[i] == 1 and not in_one:
            in_one = True
            start_t = t[i]
        # крај на сегмент (кога ќе падне на 0 или последен елемент)
        if in_one and (y_bin[i] == 0 or i == len(y_bin)-1):
            end_t = t[i] if y_bin[i] == 0 else t[i]
            ax.axvspan(start_t, end_t, alpha=alpha)
            in_one = False
```

## Sega so zero crossing rate

```python
def rolling_zcr(x, win):
    x = np.asarray(x, dtype=float)
    x = np.nan_to_num(x)

    # sign changes (length N-1)
    sc = (x[:-1] * x[1:] < 0).astype(float)

    # rolling mean over sc
    pad = win // 2
    sc_pad = np.pad(sc, (pad, pad), mode="edge")
    z = np.convolve(sc_pad, np.ones(win)/win, mode="valid")  # length N-1

    # make it length N by prepending one value (same as first)
    z = np.r_[z[0], z]   # now length N
    return z
```

```python
start_s, end_s = 220, 230
mask = (time_sec_risto >= start_s) & (time_sec_risto <= end_s)
zoom_df = df_risto.loc[mask].copy()
t = time_sec_risto[zoom_df.index].to_numpy()

fs = 50
win_s = 0.5
win_n = int(fs * win_s)

gx = zoom_df["Gyroscope X"].to_numpy()
gy = zoom_df["Gyroscope Y"].to_numpy()

zcr_gx = rolling_zcr(gx, win_n)
zcr_gy = rolling_zcr(gy, win_n)

fig, ax = plt.subplots(figsize=(12,5))

# обој Annotations==1
ann = zoom_df["Annotations"].to_numpy().astype(int)
in_one = False
for i in range(len(ann)):
    if ann[i] == 1 and not in_one:
        in_one = True
        s = t[i]
    if in_one and (ann[i] == 0 or i == len(ann)-1):
        e = t[i]
        ax.axvspan(s, e, alpha=0.15)
        in_one = False

# ZCR линии
ax.plot(t, zcr_gx, label=f"ZCR Gyro X ({win_s}s)")
ax.plot(t, zcr_gy, label=f"ZCR Gyro Y ({win_s}s)")

ax.set_xlabel("Time [s]")
ax.set_ylabel("Zero-crossing rate")
ax.set_title(f"Rolling ZCR with Annotations==1 shaded ({start_s}-{end_s}s)")
ax.legend()
plt.show()
```

* Acc Y е стабилен околу −9 m/s² → доминира гравитацијата, нема значајно вертикално движење.
* RMS Acc Y е речиси константен → без силни линеарни движења на главата.
* Acc X има кратка транзиција (~200–205 s) со зголемен RMS → промена на положба на главата / очилата.
* По транзицијата, RMS Acc X паѓа и останува низок → главата е стабилна.
* Во периодот со Annotations == 1, accelerometer сигналите се слаби и стабилни.

Chewing / swallowing активноста не генерира значајно линеарно забрзување и не е јасно видлива во accelerometer, туку главно се манифестира преку ротациони движења (gyroscope). Accelerometer служи како валидатор за стабилност на главата и за елиминација на лажни детекции.
